```python
import matplotlib.pyplot as plt
import pandas as pd

# Data from the table
data = {
    'Classifier': [
        'Random Forest',
        'Gradient Boosting',
        'Support Vector',
        'K-Nearest Neighbors',
        'Voting Method (Overfitting)'
    ],
    'Accuracy (%)': [98.65, 97.86, 79.94, 81.30, 100],
    'Precision Score': [0.99, 0.98, 0.78, 0.78, 1.00],
    'Recall Score': [0.99, 0.98, 0.82, 0.83, 1.00],
    'F1 Score': [0.99, 0.98, 0.79, 0.79, 1.00]
}

# Convert data to pandas DataFrame
df = pd.DataFrame(data)

# Plot Accuracy
plt.figure(figsize=(12, 6))
plt.bar(df['Classifier'], df['Accuracy (%)'], color='blue', alpha=0.7)
plt.title('Accuracy Comparison of Classifiers', fontsize=16)
plt.xlabel('Classifier', fontsize=14)
plt.ylabel('Accuracy (%)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Annotate the bars
for index, value in enumerate(df['Accuracy (%)']):
    plt.text(index, value + 0.5, f'{value:.2f}%', ha='center', fontsize=12)

plt.tight_layout()
plt.show()

# Plot Precision, Recall, and F1 Score Comparison
metrics = ['Precision Score', 'Recall Score', 'F1 Score']
colors = ['orange', 'green', 'purple']

plt.figure(figsize=(14, 8))

for i, metric in enumerate(metrics):
    plt.bar(
        [x + (i * 0.2) for x in range(len(df))],  # Offset for grouped bars
        df[metric],
        width=0.2,
        label=metric,
        color=colors[i],
        alpha=0.8
    )

# Customize plot
plt.title('Precision, Recall, and F1 Score Comparison', fontsize=16)
plt.xlabel('Classifier', fontsize=14)
plt.ylabel('Score', fontsize=14)
plt.xticks([x + 0.3 for x in range(len(df))], df['Classifier'], rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```
